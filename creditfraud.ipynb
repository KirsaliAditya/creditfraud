```python
import pandas as pd
```

```python
df = pd.read_csv('creditcard.csv')
df
```

```python
df.hist(bins=50, figsize=(20, 15))
```

# Preprocessing of dataset

```python
from sklearn.preprocessing import RobustScaler
new_df = df.copy()
scaler = RobustScaler()
new_df['Amount'] = scaler.fit_transform(new_df['Amount'].to_numpy().reshape(-1, 1))
time = new_df['Time']
new_df['Time'] = (time - time.min()) / (time.max() - time.min())
new_df
```

```python
new_df = new_df.sample(frac=1, random_state=42)
new_df
```

```python
train, test, val = new_df[:220000], new_df[220000:255000], new_df[255000:]
train['Class'].value_counts(), test['Class'].value_counts(), val['Class'].value_counts()
```

```python
np_train = train.to_numpy()
np_test = test.to_numpy()
np_val = val.to_numpy()

np_train.shape, np_test.shape, np_val.shape
```

```python
x_train, y_train = np_train[:, :-1], np_train[:, -1]
x_test, y_test = np_test[:, :-1], np_test[:, -1]
x_val, y_val = np_val[:, :-1], np_val[:, -1]

x_train.shape, y_train.shape, x_test.shape, y_test.shape, x_val.shape, y_val.shape
```

# Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, lr.predict(x_val),target_names=['Not fraud', 'Fraud']))
```

# Shallow neural network model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint

nn = Sequential()
nn.add(InputLayer((x_train.shape[1],)))
nn.add(Dense(2, activation='relu'))
nn.add(BatchNormalization())
nn.add(Dense(1, activation='sigmoid'))

checkpoint = ModelCheckpoint('nn.keras', save_best_only=True)
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
nn.summary()
```

```python
nn.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=10, callbacks=[checkpoint])
```

```python
def nn_predict(nn, x):
  return (nn.predict(x).flatten()>0.5).astype(int)

nn_predict(nn, x_val)
```

```python
print(classification_report(y_val, nn_predict(nn, x_val),target_names=['Not fraud', 'Fraud']))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
print(classification_report(y_val, rf.predict(x_val),target_names=['Not fraud', 'Fraud']))
```

# Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(n_estimators=50, learning_rate=1.0, max_depth=1, random_state=0)
gb.fit(x_train, y_train)
print(classification_report(y_val, gb.predict(x_val),target_names=['Not fraud', 'Fraud']))
```

Linear SVC Model

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(class_weight='balanced')
svc.fit(x_train, y_train)
print(classification_report(y_val, svc.predict(x_val),target_names=['Not fraud', 'Fraud']))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Function to plot ROC curve
def plot_roc_curve(y_true, y_pred_prob, label):
  fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob)
  roc_auc = auc(fpr, tpr)
  plt.plot(fpr, tpr, label=f'{label} (AUC = {roc_auc:.2f})')

# Get predicted probabilities for each model
y_pred_prob_lr = lr.predict_proba(x_val)[:, 1]
y_pred_prob_nn = nn.predict(x_val).flatten()
y_pred_prob_rf = rf.predict_proba(x_val)[:, 1]
y_pred_prob_gb = gb.predict_proba(x_val)[:, 1]
y_pred_prob_svc = svc.decision_function(x_val)

# Plot ROC curves for all models
plt.figure(figsize=(10, 8))
plot_roc_curve(y_val, y_pred_prob_lr, 'Logistic Regression')
plot_roc_curve(y_val, y_pred_prob_nn, 'Neural Network')
plot_roc_curve(y_val, y_pred_prob_rf, 'Random Forest')
plot_roc_curve(y_val, y_pred_prob_gb, 'Gradient Boosting')
plot_roc_curve(y_val, y_pred_prob_svc, 'Linear SVC')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Different Models')
plt.legend(loc='lower right')
plt.show()
```

# Handling Unbalanced Dataset

```python
frauds = new_df[new_df['Class']==1]
not_frauds = new_df[new_df['Class']==0]

not_frauds['Class'].value_counts(), frauds['Class'].value_counts()
```

```python
bal_df = pd.concat([frauds, not_frauds.sample(len(frauds),random_state=1)])
bal_df['Class'].value_counts()
```

```python
bal_df = bal_df.sample(frac=1, random_state=1)
bal_df
```

```python
np_bal_df = bal_df.to_numpy()

x_b_train, y_b_train = np_bal_df[:700, :-1], np_bal_df[:700, -1].astype(int)
x_b_test, y_b_test = np_bal_df[700:840, :-1], np_bal_df[700:840, -1].astype(int)
x_b_val, y_b_val = np_bal_df[840:, :-1], np_bal_df[840:, -1].astype(int)

x_b_train.shape, y_b_train.shape, x_b_test.shape, y_b_test.shape, x_b_val.shape, y_b_val.shape
```

# Picking the best model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint

b_nn = Sequential()
b_nn.add(InputLayer((x_b_train.shape[1],)))
b_nn.add(Dense(2, activation='relu'))
b_nn.add(BatchNormalization())
b_nn.add(Dense(1, activation='sigmoid'))

checkpoint = ModelCheckpoint('b_nn.keras', save_best_only=True)
b_nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
b_nn.fit(x_b_train, y_b_train, validation_data=(x_b_val, y_b_val), epochs=40, callbacks=[checkpoint])
```

```python
def b_nn_predict(nn, x):
  return (nn.predict(x).flatten()>0.5).astype(int)

b_nn_predict(b_nn, x_b_val)
```

```python
print(classification_report(y_b_val, b_nn_predict(b_nn, x_b_val),target_names=['Not fraud', 'Fraud']))
```

```python
from sklearn.svm import LinearSVC
b_svc = LinearSVC(class_weight='balanced')
b_svc.fit(x_b_train, y_b_train)
print(classification_report(y_b_val, b_svc.predict(x_b_val),target_names=['Not fraud', 'Fraud']))
```

# Classification Report on Test data

```python
print(classification_report(y_b_test, b_nn_predict(b_nn, x_b_test),target_names=['Not fraud', 'Fraud']))
```
